```python
import matplotlib.pyplot as plt
import seaborn as sns

def calcular_a1(n):
    return 1000 * n + 9000

def calcular_a2(n):
    return n**2 + 15 * n

def encontrar_n_minimo():
    n = 0
    while True:
        if calcular_a1(n) < calcular_a2(n):
            return n
        n += 1

# Encontra o valor mnimo de n em que A1  o menor
n_minimo = encontrar_n_minimo()

# Define intervalo visual apropriado
n_vals = list(range(max(0, n_minimo - 20), n_minimo + 20))
a1_vals = [calcular_a1(n) for n in n_vals]
a2_vals = [calcular_a2(n) for n in n_vals]

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

plt.plot(n_vals, a1_vals, label='A(n) = 1000n + 9000', linewidth=2)
plt.plot(n_vals, a2_vals, label='A(n) = n + 15n', linewidth=2)

# Marca o ponto de interseo
plt.axvline(x=n_minimo, color='red', linestyle='--', label=f'n = {n_minimo}')
plt.scatter([n_minimo], [calcular_a1(n_minimo)], color='red', zorder=5)

plt.title('Interseo entre A(n) e A(n)')
plt.xlabel('n')
plt.ylabel('Nmero de operaes')
plt.legend()
plt.tight_layout()
plt.show()
```
